```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Data Collection
data = {
    'Teacher': ['Teacher1', 'Teacher2', 'Teacher3', 'Teacher4'],
    'Research_Papers': [10, 15, 8, 12],
    'Attendance': [95, 85, 90, 80],
    'Seminars_Attended': [5, 7, 3, 4]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Step 3: Calculate Median
median_values = df[['Research_Papers', 'Attendance', 'Seminars_Attended']].median()

# Step 4: Track Progress
df['Above_Median_Research'] = df['Research_Papers'] > median_values['Research_Papers']
df['Above_Median_Attendance'] = df['Attendance'] > median_values['Attendance']
df['Above_Median_Seminars'] = df['Seminars_Attended'] > median_values['Seminars_Attended']

# Step 5: Visualization
# Set the index to Teacher for better visualization
df.set_index('Teacher', inplace=True)

# Create a bar plot for each metric
fig, ax = plt.subplots(3, 1, figsize=(10, 12))

# Research Papers
ax[0].bar(df.index, df['Research_Papers'], color='skyblue', label='Research Papers')
ax[0].axhline(median_values['Research_Papers'], color='red', linestyle='--', label='Median Research Papers')
ax[0].set_title('Research Papers Comparison')
ax[0].set_ylabel('Number of Research Papers')
ax[0].legend()

# Attendance
ax[1].bar(df.index, df['Attendance'], color='lightgreen', label='Attendance')
ax[1].axhline(median_values['Attendance'], color='red', linestyle='--', label='Median Attendance')
ax[1].set_title('Attendance Comparison')
ax[1].set_ylabel('Attendance Percentage')
ax[1].legend()

# Seminars Attended
ax[2].bar(df.index, df['Seminars_Attended'], color='salmon', label='Seminars Attended')
ax[2].axhline(median_values['Seminars_Attended'], color='red', linestyle='--', label='Median Seminars Attended')
ax[2].set_title('Seminars Attended Comparison')
ax[2].set_ylabel('Number of Seminars Attended')
ax[2].legend()

# Show the plot
plt.tight_layout()
plt.show()

# Step 6: Overall Summary
overall_summary = {
    'Metric': ['Average Research Papers', 'Average Attendance', 'Average Seminars Attended'],
    'Value': [
        df['Research_Papers'].mean(),
        df['Attendance'].mean(),
        df['Seminars_Attended'].mean()
    ]
}

# Create a DataFrame for the summary
summary_df = pd.DataFrame(overall_summary)

# Display the overall summary
print("\nOverall Summary:")
print(summary_df)
```
